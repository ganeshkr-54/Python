# Type Hints

```python
def get_full_name(first_name, last_name):
    full_name = first_name.title() + " " + last_name.title()
    return full_name
print(get_full_name("ganesh", "ramesh"))
```

```python
def get_full_name(first_name:str , last_name: str):
    full_name = first_name.title() + " " + last_name.title()
    return full_name
print(get_full_name("ganesh", "kumar reddy"))
```

```python
def process_iteam(iteams_t: tuple[int,int,str], iteams_s: set[bytes]):
    return iteams_s, iteams_t
print(process_iteam((3,4,"ganesh"), {b"234"}))
```

```python
def process_iteam(iteams: int | str):
    return iteams
print(process_iteam("ganesh"))
```

```python
def say_hi(name:str | None=None):
    if name:
        print(f'hello {name}!!')
    else:
        print("Hello world")
say_hi(23)
```
