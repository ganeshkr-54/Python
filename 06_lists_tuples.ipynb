## Lists

**Q&A**

Print the largest number in the list


```python
numbers = [3,5,7,4,3,9]
max = numbers[0]
for num in numbers:
    if num > max:
        max = num
print(max)
```

## 2D Lists

**Q&A**

Remove duplicates in the list

```python
numbers = [ 5,3,4,6,4,7,3]
unique = []
for num in numbers:
    if num not in unique:
        unique.append(num)
print(unique)

```
