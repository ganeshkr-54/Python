## Functions

```python
def greet():
    print("Hello Good Morning :) ")

greet()
```

## Parameters

```python
def greet(name):
    print(f"Hello Good Morning :) {name}")

greet("ganesh")
```

```python
#Positional Argument

def greet(name, pet):
    print(f"Hello {name} is your pet {pet}")
greet("ganesh",25) #here  order of position is important
```

## KeyWord Arguments

```python
#KeyWord Argument

def greet(name, pet):
    print(f"Hello {name} is your pet {pet}")
greet(pet="dog", name="ganesh") # if we intend to change order of position the we can use keyword argument
```
