```python
class Solution:

    def romanToInt(self, s: str) -> int:
        
        s = s.upper()
        roman = {
            "I" : 1,
            "V" : 5,
            "X" : 10,
            "L" : 50,
            "C" : 100,
            "D" : 500,
            "M" : 1000
    
        }
        total = 0
        p_value = 0

        for i in reversed(s):  
            value = roman[i]
            if value < p_value:
                total -= value
            else:
                total += value
            p_value = value
            


        return total



obj1=Solution()

    
print(obj1.romanToInt(s = "III"))
print(obj1.romanToInt(s = "MCMXCIV"))



```
