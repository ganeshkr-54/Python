## Dictionaries

> No duplicate key are allowed

```python
market = {
    "fruit" : "apple",
    "cost" : 50,
    "quality" : True

}

print(market["fruit"])
market["fruit"] = "Mango"
print(market["fruit"])
```

```python
phone = input("phone: ")
numbers = {
    "1" : "one" ,
    "2" : "two" ,
    "3" : "three" ,
    "4" : "four"
}
output = ''
for ch in phone:
    output += numbers.get(ch, "!") + " "
print(output)
```
