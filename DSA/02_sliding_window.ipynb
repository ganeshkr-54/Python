## Sliding Window

### Static


```python
a = [8, 3, -2, 4, 5, -1, 0, 5, 3, 9, -6]
k = 5

# Step 1: Sum of first k elements
cur = best = sum(a[:k])

# Step 2: Slide the window
for r in range(k, len(a)):
    cur = cur + a[r] - a[r - k]   # Add next element, remove element thatâ€™s left window
    best = max(best, cur)

print("Maximum sum of subarray of size", k, "=", best)
```

### Variable

```python
a = [4,5,2,0,1,8,12,3,6,9]
s=15
l,cur,best = -1,0,0

for r in range (len(a)):
    cur += a[r]
    while cur >=s:
        l += 1
        cur -= a[l]
    best = max(best,cur)

print(best)
```

```python

s = "1101100111"  # curr is the current number of zeros in the window
left = curr = ans = 0 
for right in range(len(s)):
        if s[right] == "0":
            curr += 1
        while curr > 1:
            if s[left] == "0":
                curr -= 1
            left += 1
        ans = max(ans, right - left + 1)
print(ans)  
```
