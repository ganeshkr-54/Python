Q: Price of a house is $1M.
If buyer has good credit, they need to put down payment 10% Otherwise
they need to put down payment 20%
Print the down payment

```python
price = 1000000
good_credit = True

if good_credit:
    down_payment = 0.1 * price
else: 
    down_payment = 0.2 * price

print(f"Your Down Payment is: {down_payment}")


```

## Star Patters

```python
n=5

for i in range (n): 
    for j in range(i,n):
        print("*",end=' ')
    print()
```

```python
n=5

for i in range(n):
    for j in range(i,n):
        print(" ",end=' ')
    for k in range(i+1):
        print("*",end=' ')
    print()
```

```python
n=5

for i in range(n):
    for j in range(i+1):
        print(' ', end=' ')
    for k in range(i,n):
        print("*", end=' ')
    print()
```

```python
n=5

for i in range(n):
    for j in range(i,n):
        print(" ", end=" ")
    for k in range(i+1):
        print ("*", end=' ')
    for l in range(0,i):
        print ("*", end=' ')
    print()
    
```

```python
n=5

for i in range(n):
    for j in range(i+1):
        print(" ", end=" ")
    for j in range(i,n):
        print("*", end=" ")
    for j in range(1,n-i):
        print("*", end=" ")
    print()
```
