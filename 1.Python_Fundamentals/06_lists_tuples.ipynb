## Lists

**Q&A**

Print the largest number in the list


```python
numbers = [3,5,7,4,3,9]
max = numbers[0]
for num in numbers:
    if num > max:
        max = num
print(max)
```

## 2D Lists

**Q&A**

Remove duplicates in the list

```python
numbers = [ 5,3,4,6,4,7,3]
unique = []
for num in numbers:
    if num not in unique:
        unique.append(num)
print(unique)

```

## Sorting

sort(): to sort list of iteams "permanently" 


sort(reverse=True): to reverse list of iteams "permanently"


sorted(): to sort temporarily

reverse(): To reverse the list of iteam without any specific rule.

```python
vehicles = ['car', 'bike', 'cycle']
vehicles.sort()  
print(vehicles)  
vehicles.sort(reverse=True)
print(vehicles)
print(sorted(vehicles))
print(vehicles)  
len(vehicles)

```

```python
places = ["gujarat", "telangana", "andhra", "kerala", "jammu"]

print(places)
print(sorted(places))
print(places)
print(sorted(places, reverse=True))
print(places)
places.reverse()
print(places)
places.reverse()
print(places)
places.sort()
print(places)
```

## Coping a List

```python
list1 = [1,2,3,4]

list = list1[:]  #1, 

print(list)
list1.append(5) #if you append a new element to the parent list it wont add to child list
print(list) # it is not priting 5
print(list1)# it is priting 5

list = list1  #2 
list1.append(9)   #if you append a new element to the parent list it will add to child list as well
print(list) # printing 9
print(list1) # printing 9
```

```python
list = [1,2,3,4,5,6,7,8,9]
mid = len(list)//2
print(list[mid-1:mid+2]) # here we used +2 instead +2 because Python slicing includes start but excludes end

print(list[-3:]) # last 3 elements
```
