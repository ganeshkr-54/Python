### Object


An object is a specific instance of a class. It holds its own set of data (instance variables) and can invoke methods defined by its class. Multiple objects can be created from same class, each with its own unique attributes.

Let's create an object from Dog class.




````python
class Computer:
    def __init__(self,cpu,ram):    # automatically initialize the variables
        self.cpu = cpu
        self.ram = ram
    def config(self):
        print("Config is", self.cpu, self.ram)

com1 = Computer("i5", 16) #by default "self" is also passed along with parameters
com2 = Computer("Ryzen 3", 8)

com1.config()
com2.config()


        
````

## Accessing Class and Instance attributes

````python
class India:
    place = "srisailam"

    def __init__(self, mountain, river):
        self.mountain = mountain
        self.river = river
    def pride(self):
        print(f"india has {self.mountain} and {self.river}")

you = India("sahyadri", "Godavari") # creating object of the class Inida
you.pride() # here we are calling instance(method) variables
print(you.place)   # Calling class variables
        
````

## Getter and Setter


They are methods that control how you access (get) and modify (set) an object‚Äôs attributes.

Getter ‚Üí used to read a private or protected variable.

Setter ‚Üí used to modify it safely, often with validation logic.

They let you enforce rules or hide internal details (called encapsulation).

üêç Python‚Äôs Special Way ‚Äì @property

Unlike Java or C++, in Python you don‚Äôt need to write .getName() or .setName().
Python gives a clean, elegant syntax using decorators:



````python
class Dog:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self,value):
        if value <0:
            print("negivite number")
        else:
            self._age = value
    def nname(self):
        print(f"hello my new dog name is {self._name} and its age is {self._age}")
        #return f"hello my new dog name is {self._name} and its age is {self._age}"

obj1=Dog("Tommy",25)
print(obj1.name)
print(obj1.age)
obj1.name = "chimtu"
obj1.age = 15
obj1.nname()

        
````

## Method Overriding

````python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):  # Method overriding
        print("Woof")

dog = Dog()
dog.sound()
````
