# Python's Instance, Class, and Static Methods

Instance methods access the state of a specific object through the self parameter.\
 You create class methods with the @classmethod decorator and use them for operations that involve class-level data.\
 You use static methods for utility functionality that doesn’t need class or instance data, and you create them with the @staticmethod decorator.\
 Using class methods and static methods in your classes can improve class design and code maintainability.


````python
class employee:                                   # class
    amount_raise = 1.07    # Class Variable
    no_of_emply = 0
    def __init__(self, first, last, salary):        #method or # function
        self.first = first
        self.last = last
        self.salary = salary
        employee.no_of_emply += 1

    def fullname(self):
        return self.first + ' ' + self.last
    def apply_raise(self):
        self.salary = int(self.salary * self.amount_raise)
    
    @classmethod
    def set_raise_amount(cls, amount):
        cls.amount_raise = amount
    
    @staticmethod
    def is_workday(day):
        if day.weekday == 5 or day.weekday ==6:
            return False
        return True

emp1 = employee("ganesh","reddy",20)    # instance
emp2 = employee("shiva","om", 40)       # instance

print(employee.amount_raise)
print(emp1.amount_raise)
print(emp2.amount_raise)

import datetime
my_date = datetime.date(2016, 7, 10)
print(employee.is_workday(my_date))
````

## Encapsulation

Encapsulation is about protecting data inside a class.

It means keeping data (properties) and methods together in a class, while controlling how the data can be accessed from outside the class.

This prevents accidental changes to your data and hides the internal details of how your class works.

Private Properties
In Python, you can make properties private by using a double underscore __ prefix:


#### Encapsulation provides several benefits:

Data Protection: Prevents accidental modification of data


Validation: You can validate data before setting it


Flexibility: Internal implementation can change without affecting external code


Control: You have full control over how data is accessed and modified

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self._age = age

p1=Person("ganesh", 25)
print(p1.name)
print(p1._age)


    
````

## Getter and Setter


They are methods that control how you access (get) and modify (set) an object’s attributes.

Getter → used to read a private or protected variable.

Setter → used to modify it safely, often with validation logic.

They let you enforce rules or hide internal details (called encapsulation).

🐍 Python’s Special Way – @property

Unlike Java or C++, in Python you don’t need to write .getName() or .setName().
Python gives a clean, elegant syntax using decorators:



````python
class Dog:
    def __init__(self, name, age):
        self._name = name  # _(underscore) in the variables indicates they are private
        self._age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self,value):
        if value <0:
            print("negivite number")
        else:
            self._age = value
    def nname(self):
        print(f"hello my new dog name is {self._name} and its age is {self._age}")
        #return f"hello my new dog name is {self._name} and its age is {self._age}"

obj1=Dog("Tommy",25)
print(obj1.name)
print(obj1.age)
obj1.name = "chimtu"
obj1.age = 15
obj1.nname()

        
````

## Method Overriding

````python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):  # Method overriding
        print("Woof")

dog = Dog()
dog.sound()
````

### Self vs Super

````python
class School:
    def __init__(self,name,age,degree):
        self.name = name
        self.degree = degree
        self.age = age
    
class Student(School):
    def __init__(self, name, age, degree,roll_number):
        super().__init__(name, age, degree) # helps to remove duplication of code, in future if parent class is updated with new variables it will automatically initialize into child class as well
        self.roll_number = roll_number
````
