```python
def is_valid(s: str) -> bool:
    stack = []
    bracket_map = {")": "(", "}": "{", "]": "["}

    for char in s:
        if char in bracket_map:  # Closing bracket
            if stack and stack[-1] == bracket_map[char]:
                stack.pop()
            else:
                return False
        else:  # Opening bracket
            stack.append(char)

    return not stack
```
